## 2. Analyse score statistics

```python
area_group = df.groupby('shot_zone_area')
basic_group = df.groupby('shot_zone_basic')
range_group = df.groupby('shot_zone_range')
```

```python
plt.subplots(1, 3, figsize=(14, 7), sharey=True)
plt.title('All shot made in his career')
colors = list('rgbcmyk')

plt.subplot(131)
plt.ylim(800, -50)
plt.title('shot_zone_area')
for i, (_, area) in enumerate(area_group):
    plt.scatter(area['loc_x'], area['loc_y'], alpha=0.1, color=colors[i])
    
plt.subplot(132)
plt.ylim(800, -50)
plt.title('shot_zone_basic')
for i, (_, basic) in enumerate(basic_group):
    plt.scatter(basic['loc_x'], basic['loc_y'], alpha=0.1, color=colors[i])

    
plt.subplot(133)
plt.ylim(800, -50)
plt.title('shot_zone_range')
for i, (_, shot_zone_range) in enumerate(range_group):
    plt.scatter(shot_zone_range['loc_x'], shot_zone_range['loc_y'], alpha=0.1, color=colors[i])
```

#### 每一年，科比投篮区域的变化

```python
years = 21
rows = 7
columns = 3

plt.subplots(rows, columns, figsize=(14,7*rows), sharey=True)
plt.title('All shot made each year in his career')
colors = list('rgbcmyk')

for i in range(years):
    plt.subplot(rows, columns, i+1)
    plt.ylim(800, -50)
    yearly_data = df[df['game_year'] == (1996+i)]
    plt.title('Year {}, total shots {}'.format(1996+i, len(yearly_data)))
    for i, (_, area) in enumerate(yearly_data.groupby('shot_zone_area')):
        plt.scatter(area['loc_x'], area['loc_y'], alpha=0.1, color=colors[i])
```

#### 每一年，科比得分方式的变化

```python
years = 21
rows = 7
columns = 3

plt.subplots(rows, columns, figsize=(14,7*rows), sharey=True)
plt.title('All shot made each year in his career')
colors = list('rgbcmyk')

for i in range(years):
    plt.subplot(rows, columns, i+1)
    plt.ylim(800, -50)
    yearly_data = df[df['game_year'] == (1996+i)]
    shot_types = yearly_data['combined_shot_type'].unique()
#     print 1996+i, shot_types
    plt.title('Year {}, shot types {}'.format(1996+i, len(shot_types)))
    for j, (_, area) in enumerate(yearly_data.groupby('combined_shot_type')):
        plt.scatter(area['loc_x'], area['loc_y'], alpha=0.1, color=colors[j])
```
